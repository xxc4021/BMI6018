```python
import numpy as np
import pandas as pd
```

```python
diabetic_data = pd.read_csv('diabetic_data.csv')
diabetic_data.head()
```

```python
q_1 = diabetic_data[['encounter_id','gender','admission_type_id']]
q_1_answer = pd.melt(q_1, id_vars=['encounter_id'], value_vars=['admission_type_id'])
q_1_answer
#show the admission_type_id value 
```

```python
q_2 = diabetic_data[['encounter_id','admission_type_id','discharge_disposition_id']]
q_2_answer = q_2.pivot(index='encounter_id', columns='admission_type_id', values='discharge_disposition_id')
q_2_answer
#show the admission_type_id and discharge_disposition_id of different patients
```

```python
q_3 = diabetic_data[['num_lab_procedures','num_procedures']]
q_3_answer = q_3.agg("mean")
q_3_answer
#calculate the mean numbers of lab procedures and procedures
```

```python
q_4= diabetic_data[['encounter_id','age']]
for i,l in q_4.iterrows(): 
    print(l['encounter_id'],l['age']) 
#print the encounter_id and age range 
```

```python
q_5= diabetic_data[['age','num_lab_procedures']]
q_5_answer = q_5.groupby(by = 'age').mean()
q_5_answer
#calculate the mean numbers of lab procedures group by age
```
